```python
# Importing packages.

import pandas as pd
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt

%matplotlib inline 

# Bringing in the data.

df = pd.read_csv(r'//Users/mirnessabanovic/Downloads/fetal_health.csv')
```

```python
# Initial look at the data.

df.head()
```

```python
# Checking the data for empty/missing values.
# Using a for loop to check percentage of data missing for each column. 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# No missing or null values.
```

```python
# Finding the data type for every column.

print(df.dtypes)
```

```python
# Checking for outliers in the baseline fetal heart rate column.

df.boxplot(column=['baseline value'])
```

```python
# Checking for duplicate rows.

df[df.duplicated()]
```

```python
# Deleting duplicate rows.

df.drop_duplicates()
```

```python
# Now, I will begin to look at what factors are most correlated with poor fetal health outcomes.

# I predict that pregnancies that have a high abnormal short-term and long-term heart rate variability 
# are also more likely to give birth to fetuses with pathological conditions.
# I also predict that fetuses that have high short-term and long-term heart rate variability will also have
# high baseline fetal heart rates.

# Scatter plot of short-term heart rate variability and baseline heart rate.

plt.scatter(x = df['abnormal_short_term_variability'], y = df ['baseline value'])

plt.title('Short Term Heart Rate Variability vs Baseline Heart Rate')

plt.xlabel('Short Term Heart Rate Variability')

plt.ylabel('Baseline Heart Rate')

plt.show()
```

```python
# Scatter plot of long-term heart rate variability and baseline heart rate.

plt.scatter(x = df['percentage_of_time_with_abnormal_long_term_variability'], y = df ['baseline value'])

plt.title('Long Term Heart Rate Variability vs Baseline Heart Rate')

plt.xlabel('Long Term Heart Rate Variability')

plt.ylabel('Baseline Heart Rate')

plt.show()
```

```python
# Here, I will use a regression plot to visualize the correlation between short-term heart rate variability 
# and baseline fetal heart rate.

sns.regplot(x = 'abnormal_short_term_variability', y ='baseline value', 
            data = df, scatter_kws = {"color": "red"}, line_kws = {"color":"blue"})
```

```python
# A regression plot to visualize the correlation between long-term heart rate variability 
# and baseline fetal heart rate.

sns.regplot(x = 'percentage_of_time_with_abnormal_long_term_variability', y ='baseline value',
            data = df, scatter_kws = {"color": "red"}, line_kws = {"color":"blue"})
```

```python
# There is a slight positive correlation between both short-term and long-term heart rate variability and
# baseline fetal heart rate.
# Here, I will create a correlation matrix to observe the relationship between all the variables in the table to 
# fetal health outcomes.

df.corr(method = 'pearson')
```

```python
# Both short-term and long-term heart rate variability have a moderate correlation with fetal health outcomes.
# My prediction was correct.
```
